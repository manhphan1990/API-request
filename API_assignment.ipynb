## Normalize the data

```python
# Normalize the Hashtag_to_media post
df=pd.json_normalize(data['body']['edge_hashtag_to_media']['edges'])
df
```

```python
# Normalize the hashtag_to_top_posts
df1=pd.json_normalize(data['body']['edge_hashtag_to_top_posts']['edges'])
df1
```

```python
df1.shape
```

## Tidying

```python
# Since there are different number in the columns of each dataset, so i will take what i need in common
print(df.columns)
print(df1.columns)
```

```python
# Take what i need from 1st dataset
new_df=df[['node.id','node.accessibility_caption','node.taken_at_timestamp','node.edge_liked_by.count',
           'node.edge_media_to_comment.count','node.shortcode','node.owner.id','node.is_video']]
new_df
```

```python
# Take what i need from 2nd dataset 
new_df1=df1[['node.id','node.accessibility_caption','node.taken_at_timestamp','node.edge_liked_by.count',
             'node.edge_media_to_comment.count','node.shortcode','node.owner.id','node.is_video']]
new_df1
```

```python
print(new_df.shape)
print(new_df1.shape)
```

```python
# Both dataset should have same columns now, it time to merge them
Clean_df=pd.concat([new_df,new_df1],ignore_index=True)
```

```python
Clean_df
```

```python
# Adding a column with the text in post
Post_text=[]
for i in data['body']['edge_hashtag_to_media']['edges']:
    Post_text.append(i['node']['edge_media_to_caption']['edges'][0]['node']['text'])
for j in data['body']['edge_hashtag_to_top_posts']['edges']:
    Post_text.append(j['node']['edge_media_to_caption']['edges'][0]['node']['text'])
print(Post_text)
len(Post_text)
```

```python
post=pd.DataFrame(Post_text,columns=['Post_text'])
post
```

```python
Clean_df=pd.concat([Clean_df,post],axis=1)
```

```python
Clean_df.shape
```

```python
# Changing timestamp to datetime
Clean_df['node.taken_at_timestamp'] = pd.to_datetime(Clean_df['node.taken_at_timestamp'], unit='s')
```

```python
Clean_df['node.taken_at_timestamp']
```

```python
Clean_df.columns
```

```python
# Rename the default column names to understandable names
Clean_df.rename(columns={'node.id':'Post_ID',
                         'node.accessibility_caption': 'Post_caption',
                         'node.taken_at_timestamp': 'Post_Date',
                         'node.edge_liked_by.count': 'Likes_count',
                         'node.edge_media_to_comment.count': 'Post_comment_count',
                         'node.shortcode': 'Post_shortcode',
                         'node.owner.id': 'Owner_ID',
                         'node.is_video': 'Video_type'},
                inplace=True)
```
